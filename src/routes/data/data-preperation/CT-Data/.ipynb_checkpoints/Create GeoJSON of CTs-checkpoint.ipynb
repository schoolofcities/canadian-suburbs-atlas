```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Polygon
```

```python
classif = gpd.read_file("ct-classification-data.csv")
shape = gpd.read_file("ct.geojson")
```

```python
merge = shape.merge(classif, left_on="ctuid", right_on = 'CTUID', how = 'left')
sub = merge[["CMA" , "CTUID" , "CLASS" , "POP2021" , "DU2021" , "TOTAL_COM" , "ACTIVE" , "TRANSIT" , "AUTO" , "geometry_x" , "landarea"]]
ct_final = sub.rename(columns = {"geometry_x" : "geometry"})
ct_final
```

```python
ct_final.plot()
```

```python
gdf = gpd.GeoDataFrame(ct_final, geometry='geometry')
gdf.plot()
```
